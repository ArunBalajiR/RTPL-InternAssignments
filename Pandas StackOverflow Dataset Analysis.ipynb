## Filtering

```python
df.head()
```

```python
#Filtering people having older than 20
age = df['Age'] > 20
df.loc[age]
```

```python
#filtering people older than 20
age= (df['Age']>20)
df.loc[age,'Country']  #filter only these columns 
```

```python
#people older than 20 with these columns
df.loc[age,['Country','LanguageWorkedWith','Age1stCode','YearsCode']]
```

```python
#I only want these list of people from below countries
countries = ['India','United States','United Kingdom','France','Brazil']
filt = df['Country'].isin(countries)  #note this isin method
df.loc[filt]
```

```python
df.loc[filt,'Country']
```

```python
df['LanguageWorkedWith']
```

```python
#If we notice here, we can see each programming lang is seperated by ;
filt = df['LanguageWorkedWith'].str.contains('Python',na=False)  #na=False omit null values
df.loc[filt,'LanguageWorkedWith']
```
